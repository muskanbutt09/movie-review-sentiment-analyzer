```python
# ====================== AI RESEARCH PAPER SUMMARIZER - TASK 3 ======================
# Author: Ready for submission | Uses only allowed libraries (nltk + sklearn)

import nltk
import re
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer

# One-time setup (runs automatically)
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('punkt_tab', quiet=True) # Added to resolve the LookupError

def summarize_paper(text):
    if not text.strip():
        return "No text provided."

    # Step 1: Split into sentences
    sentences = sent_tokenize(text)
    if len(sentences) < 3:
        return "Text too short for summarization."

    # Step 2: Preprocessing
    stop_words = set(stopwords.words('english'))
    cleaned_sentences = []
    for sent in sentences:
        # Lowercase + remove punctuation + remove stopwords
        clean = re.sub(r'[^a-zA-Z0-9\s]', '', sent.lower())
        words = [word for word in word_tokenize(clean) if word not in stop_words]
        cleaned_sentences.append(' '.join(words))

    # Step 3: TF-IDF Scoring for summary
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(cleaned_sentences)
    sentence_scores = np.array(tfidf_matrix.sum(axis=1)).flatten()

    # Get top 4 sentences for summary (or less if short text)
    top_n = min(4, len(sentences))
    top_indices = sentence_scores.argsort()[-top_n:][::-1]
    summary_sentences = [sentences[i] for i in sorted(top_indices)]
    short_summary = ' '.join(summary_sentences[:2])  # Keep summary very concise

    # Step 4: Extract Key Points (top 3-5 sentences)
    key_points = [f"- {sentences[i]}" for i in sorted(top_indices)[:5]]

    # Step 5: Keyword Extraction (top 5 important words)
    feature_names = vectorizer.get_feature_names_out()
    word_scores = np.array(tfidf_matrix.sum(axis=0)).flatten()
    top_word_indices = word_scores.argsort()[-8:][::-1]  # 8 to pick meaningful ones
    keywords = [feature_names[i] for i in top_word_indices if len(feature_names[i]) > 3][:5]

    # Step 6: Conclusion (last sentence or highest scoring sentence near end)
    conclusion = sentences[-1]
    if len(sentences) > 3:
        # Bonus: pick the highest scoring sentence from last third
        last_third = sentence_scores[len(sentences)//3*2:]
        if len(last_third) > 0:
            best_concl_idx = len(sentences)//3*2 + np.argmax(last_third)
            conclusion = sentences[best_concl_idx]

    return {
        'summary': short_summary,
        'key_points': key_points,
        'keywords': keywords,
        'conclusion': conclusion
    }

# ====================== MAIN PROGRAM ======================
print("🤖 AI Research Paper Summarizer – Advanced Task 3")
print("=" * 70)
print("Paste your research paper / article text below.")
print("When finished, type 'END' on its own line and press Enter.\n")

lines = []
while True:
    try:
        line = input("> ")
        if line.strip().upper() == "END":
            break
        lines.append(line)
    except:
        break

full_text = "\n".join(lines)

if not full_text.strip():
    print("No text entered.")
else:
    print("\n🔍 Analyzing paper...\n")
    result = summarize_paper(full_text)

    print("📝 SUMMARY:")
    print(result['summary'])
    print("\n🔑 KEY POINTS:")
    for point in result['key_points']:
        print(point)
    print("\n🏷️  KEYWORDS:")
    print(", ".join(result['keywords']))
    print("\n📌 CONCLUSION:")
    print(result['conclusion'])

    print("\n" + "="*70)
    print("✅ Task 3 completed! Ready for submission.")
```
