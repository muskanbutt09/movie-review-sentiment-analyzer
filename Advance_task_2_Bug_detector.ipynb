```python
import ast

def detect_and_fix_bugs(user_code):
    issues = []
    lines = user_code.splitlines(keepends=True)  # keep exact lines for correction

    # 1. Check syntax with Python's AST (catches missing :, parentheses, indentation, etc.)
    try:
        ast.parse(user_code)
    except SyntaxError as e:
        issue = {
            'type': 'Syntax Error',
            'line': e.lineno,
            'message': e.msg,
            'explanation': '',
            'suggested_fix': '',
            'corrected_example': ''
        }

        # Specific, user-friendly explanations (covers 4 common syntax bugs)
        if e.msg == "expected ':'":
            issue['explanation'] = "In Python, 'for', 'if', 'while', 'def', 'class', etc. MUST end with a colon (:)."
            issue['suggested_fix'] = f"Add ':' at the end of line {e.lineno}."
            if e.lineno and e.lineno <= len(lines):
                original = lines[e.lineno-1].rstrip('\n')
                corrected = original + ':'
                issue['corrected_example'] = f"Corrected line {e.lineno}:\n{corrected}"

        elif "'(" in e.msg and "never closed" in e.msg:
            issue['explanation'] = "Parentheses are not balanced – you opened '(' but never closed it with ')'."
            issue['suggested_fix'] = "Add the missing closing parenthesis."
            if e.lineno and e.lineno <= len(lines):
                original = lines[e.lineno-1].rstrip('\n')
                missing = original.count('(') - original.count(')')
                corrected = original + ')' * missing
                issue['corrected_example'] = f"Corrected line {e.lineno}:\n{corrected}"

        elif "unexpected indent" in e.msg.lower():
            issue['explanation'] = "Extra indentation where there should be none."
            issue['suggested_fix'] = f"Remove the spaces/tabs at the start of line {e.lineno}."
            if e.lineno and e.lineno <= len(lines):
                corrected = lines[e.lineno-1].lstrip()
                issue['corrected_example'] = f"Corrected line {e.lineno}:\n{corrected}"

        elif "expected an indented block" in e.msg:
            issue['explanation'] = "After the colon (:), the next line MUST be indented (4 spaces)."
            issue['suggested_fix'] = f"Indent line {e.lineno} with 4 spaces."
            issue['corrected_example'] = "Example:\nfor i in range(5):\n    print(i)  # ← indented"

        else:
            issue['explanation'] = f"General syntax error: {e.msg}"
            issue['suggested_fix'] = f"Check line {e.lineno} for missing quotes, commas, or typos."

        issues.append(issue)

    else:
        # 2. If syntax is OK, check for runtime errors (e.g. undefined variables)
        try:
            # Safe exec (no dangerous built-ins)
            exec(user_code, {"__builtins__": {}}, {})
        except NameError as e:
            issues.append({
                'type': 'NameError (Undefined Variable)',
                'line': 'N/A (runtime)',
                'explanation': str(e) + ". You used a name that doesn't exist yet.",
                'suggested_fix': "Define the variable before using it (or fix spelling).",
                'corrected_example': 'Example fix:\nx = 10\nprint(x)'
            })
        except Exception as e:
            issues.append({
                'type': type(e).__name__,
                'explanation': str(e),
                'suggested_fix': "Fix the logic or calculation error.",
                'corrected_example': 'Run the code step-by-step to debug.'
            })

    return issues


# ====================== MAIN PROGRAM ======================
print("🤖 AI Bug Detection Assistant – Advanced Task 2")
print("=" * 65)
print("Enter your Python code line by line.")
print("When finished, type 'END' on its own line and press Enter.\n")

lines = []
while True:
    try:
        line = input("> ")
        if line.strip().upper() == "END":
            break
        lines.append(line)
    except:
        break

user_code = "\n".join(lines)

if not user_code.strip():
    print("No code entered.")
else:
    print("\n🔍 Analyzing...\n")
    bugs = detect_and_fix_bugs(user_code)

    if bugs:
        for i, bug in enumerate(bugs, 1):
            print(f"🚨 Detected Issue #{i}: {bug['type']}")
            if bug.get('line') and bug['line'] != 'N/A (runtime)':
                print(f"   Line: {bug['line']}")
            print(f"   Explanation: {bug['explanation']}")
            print(f"   Suggested Fix: {bug['suggested_fix']}")
            if bug.get('corrected_example'):
                print(f"   Corrected Example:\n{bug['corrected_example']}")
            print("-" * 65)
    else:
        print("✅ No common bugs detected! Excellent code!")

    print("\n💡 Best practice tip: Always use 4 spaces for indentation.")
```
