```python
# ====================== AI INTERN TASK 4 - PROFESSIONAL LEARNING PATH ENGINE ======================
# Author: Professional Version | Uses only pandas + built-in Python
# Features: Domain detection, Level classification, Structured roadmap, Clean output

import pandas as pd
import re

# ====================== PROFESSIONAL KNOWLEDGE BASE (pandas DataFrame) ======================
# This is the "database" of all learning paths - easy to extend

data = {
    'Domain': ['AI', 'Web Development', 'Data Science', 'Cybersecurity'],
    'Keywords': [
        ['ai', 'artificial intelligence', 'machine learning', 'deep learning', 'neural'],
        ['web', 'html', 'css', 'javascript', 'react', 'frontend', 'backend'],
        ['data science', 'pandas', 'numpy', 'statistics', 'visualization', 'sql'],
        ['cybersecurity', 'hacking', 'security', 'penetration', 'cryptography', 'network']
    ],
    'Beginner_Path': [
        "1. Python Fundamentals\n2. Mathematics for AI\n3. Introduction to Machine Learning\n4. Build First Project (Chatbot)",
        "1. HTML & CSS Basics\n2. JavaScript Fundamentals\n3. Build Responsive Website\n4. Introduction to React",
        "1. Python for Data\n2. Data Analysis with Pandas & NumPy\n3. Data Visualization\n4. SQL Basics",
        "1. Networking Basics\n2. Linux Commands\n3. Introduction to Cybersecurity\n4. Ethical Hacking Fundamentals"
    ],
    'Intermediate_Path': [
        "1. Advanced Python & OOP\n2. Deep Learning with TensorFlow/Keras\n3. Computer Vision / NLP\n4. Deploy AI Models",
        "1. Advanced JavaScript\n2. Backend with Node.js or Django\n3. Database (MongoDB/PostgreSQL)\n4. Full-Stack Project",
        "1. Statistics & Probability\n2. Machine Learning Algorithms\n3. Big Data Tools (Spark)\n4. End-to-End Data Pipeline",
        "1. Advanced Network Security\n2. Penetration Testing Tools\n3. Cryptography Deep Dive\n4. CTF Challenges"
    ],
    'Advanced_Path': [
        "1. Research Papers & Latest Papers\n2. MLOps & Model Deployment\n3. AI Ethics & Responsible AI\n4. Contribute to Open Source AI",
        "1. Microservices Architecture\n2. Performance Optimization\n3. DevOps & CI/CD\n4. Build Scalable SaaS Product",
        "1. Deep Learning & Research\n2. Time Series & Advanced Analytics\n3. Data Engineering\n4. Lead Data Science Team Project",
        "1. Red Team / Blue Team Operations\n2. Malware Analysis\n3. Cloud Security (AWS/Azure)\n4. Security Architecture Design"
    ],
    'Topics': [
        ['Python', 'NumPy', 'Pandas', 'Scikit-learn', 'TensorFlow'],
        ['HTML', 'CSS', 'JavaScript', 'React', 'Node.js'],
        ['Python', 'Pandas', 'SQL', 'Tableau', 'Machine Learning'],
        ['Networking', 'Linux', 'Metasploit', 'Wireshark', 'Cryptography']
    ],
    'Courses': [
        ['Python for Everybody (Coursera)', 'Machine Learning by Andrew Ng (Coursera)', 'Deep Learning Specialization'],
        ['The Web Developer Bootcamp (Udemy)', 'React - The Complete Guide (Udemy)', 'Node.js Course'],
        ['Data Science Specialization (Coursera)', 'Python for Data Analysis', 'SQL for Data Science'],
        ['The Complete Ethical Hacking Course (Udemy)', 'Certified Ethical Hacker (CEH)', 'CompTIA Security+']
    ]
}

kb = pd.DataFrame(data)

# ====================== CORE FUNCTIONS ======================
def detect_level(text):
    text_lower = text.lower()
    if any(word in text_lower for word in ['beginner', 'new', 'start', 'no experience']):
        return "Beginner"
    elif any(word in text_lower for word in ['intermediate', 'medium', 'some experience']):
        return "Intermediate"
    elif any(word in text_lower for word in ['advanced', 'expert', 'pro', 'senior']):
        return "Advanced"
    return "Beginner"  # safe default

def detect_domain(text):
    text_lower = text.lower()
    max_score = 0
    best_domain = "AI"  # default

    for idx, keywords in enumerate(kb['Keywords']):
        score = sum(1 for kw in keywords if kw in text_lower)
        if score > max_score:
            max_score = score
            best_domain = kb.loc[idx, 'Domain']
    return best_domain

def get_recommendation(user_text):
    level = detect_level(user_text)
    domain = detect_domain(user_text)

    # Get row
    row = kb[kb['Domain'] == domain].iloc[0]

    # Select path based on level
    if level == "Beginner":
        path = row['Beginner_Path']
    elif level == "Intermediate":
        path = row['Intermediate_Path']
    else:
        path = row['Advanced_Path']

    topics = row['Topics']
    courses = row['Courses']

    return {
        'level': level,
        'domain': domain,
        'path': path,
        'topics': topics,
        'courses': courses
    }

# ====================== MAIN PROFESSIONAL INTERFACE ======================
print("🤖 AI Personalized Learning Path Engine – Advanced Task 4")
print("=" * 75)
print("Paste your skills, interests & goals below (e.g. 'I am a beginner and I want to learn Artificial Intelligence.')")
print("When finished, type 'END' on its own line and press Enter.\n")

lines = []
while True:
    try:
        line = input("> ")
        if line.strip().upper() == "END":
            break
        lines.append(line)
    except:
        break

user_input = "\n".join(lines)

if not user_input.strip():
    print("No input provided.")
else:
    print("\n🔍 Analyzing your profile...\n")
    rec = get_recommendation(user_input)

    print(f"👤 User Level: {rec['level']}")
    print(f"🎯 Recommended Domain: {rec['domain']}")
    print("\n📚 RECOMMENDED LEARNING PATH:")
    print(rec['path'])
    print("\n🏷️  SUGGESTED TOPICS:")
    for t in rec['topics']:
        print(f"• {t}")
    print("\n📖 RECOMMENDED COURSES:")
    for c in rec['courses']:
        print(f"• {c}")

    print("\n" + "="*75)
    print("✅ Professional Recommendation Generated! Ready for submission.")
```
